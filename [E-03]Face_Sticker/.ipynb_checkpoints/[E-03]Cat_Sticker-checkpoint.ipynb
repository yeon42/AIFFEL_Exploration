# 1. ÏÇ¨ÏßÑ Ï§ÄÎπÑÌïòÍ∏∞

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
my_image_path = os.getenv('HOME') + '/aiffel/camera_sticker/images/me_front.png'
img_bgr = cv2.imread(my_image_path) # OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
img_show = img_bgr.copy() # Ï∂úÎ†• Ïù¥ÎØ∏ÏßÄ Îî∞Î°ú Î≥¥Í¥Ä
plt.imshow(img_bgr)
plt.show()
```

üîÆÏùºÎ∞òÏ†ÅÏù∏ Ïù¥ÎØ∏ÏßÄ ÎùºÏù¥Î∏åÎü¨Î¶¨Îì§ÏùÄ Î™®Îëê Ïù¥ÎØ∏ÏßÄÎ•º RGB ÏàúÏúºÎ°ú ÏÇ¨Ïö©ÌïòÏßÄÎßå <br/>
üîÆopencvÎäî ÏòàÏô∏Ï†ÅÏúºÎ°ú BGRÏùÑ ÏÇ¨Ïö©ÌïúÎã§. <br/>
‚ñ∂Ô∏è Ïù¥ ÎïåÎ¨∏Ïóê ÏÉâÍπîÏù¥ Î∞òÏ†ÑÎêòÏñ¥ ÎÇòÏò§Îäî Í≤É!

```python
# COLOR_BGR2RGB: BGR Ïù¥ÎØ∏ÏßÄÎ•º RGB Ïù¥ÎØ∏ÏßÄÎ°ú Î∞îÍæ∏Í∏∞
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

# 2. ÏñºÍµ¥ Í≤ÄÏ∂ú Face Detection

dlibÏùò face detectorÏùÄ HOG(Histogram of Oriented Gradients)ÏôÄ SVM(Support Vector Machine)ÏùÑ ÏÇ¨Ïö©Ìï¥ ÏñºÍµ¥ÏùÑ Ï∞æÎäîÎã§. <br/>

### üîÆHOGÎûÄ?
Ïù¥ÎØ∏ÏßÄÏùò ÏÉâÏÉÅ Î≥ÄÌôîÎüâÏùÑ ÎÇòÌÉÄÎÉÑ -> Í∞ùÏ≤¥(ÏÇ¨Îûå, ÏÇ¨Î¨º)Ïùò ÌòïÌÉú Ï∞æÏùÑ Ïàò ÏûàÏùå
### üîÆSVMÏù¥ÎûÄ?
ÏÑ†Ìòï Î∂ÑÎ•òÍ∏∞ -> Ïó¨Í∏∞ÏÑ† ÏñºÍµ¥Ïù∏ÏßÄÎ•º Íµ¨Î∂ÑÌïòÍ≤†ÏßÄ!

```python
# detector ÏÑ†Ïñ∏
detector_hog = dlib.get_frontal_face_detector()

# detector_hogÎ•º Ïù¥Ïö©Ìï¥ ÏñºÍµ¥Ïùò bounding box Ï∂îÏ∂ú
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1) # (image, # of image pyramid)
```

-> dlibÏùÄ rgb Ïù¥ÎØ∏ÏßÄÎ•º ÏûÖÎ†•ÏúºÎ°ú Î∞õÍ∏∞ ÎïåÎ¨∏Ïóê cvtColorÏùÑ Ïù¥Ïö©Ìï¥ Î≥ÄÌòïÌï¥Ï§òÏïºÌïúÎã§!

```python
# Ï∞æÏùÄ ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ Î¶¨Ïä§Ìä∏
print(dlib_rects)

for dlib_rect in dlib_rects:
    l = dlib_rect.left() # ÏôºÏ™Ω ÏïÑÎûò
    t = dlib_rect.top() # ÏôºÏ™Ω ÏúÑ
    r = dlib_rect.right() # Ïò§Î•∏Ï™Ω ÏúÑ
    b = dlib_rect.bottom() # Ïò§Î•∏Ï™Ω ÏïÑÎûò
    
    cv2.rectangle(img_show, (l, t), (r, b), (0, 255, 0), 2, lineType=cv2.LINE_AA)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

‚úîÔ∏ècv2.rectangel(Ïù¥ÎØ∏ÏßÄ ÌååÏùº, ÏãúÏûëÏ†ê Ï¢åÌëú, Ï¢ÖÎ£åÏ†ê Ï¢åÌëú, ÏÉâÏÉÅ, ÏÑ† ÎëêÍªò, ÏÑ† Ï¢ÖÎ•ò)

# 3. ÏñºÍµ¥ ÎûúÎìúÎßàÌÅ¨ Face Landmark

### üîÆface landmark localization
: Ïù¥Î™©Íµ¨ÎπÑÏùò ÏúÑÏπòÎ•º Ï∂îÎ°†ÌïòÎäî Í≤É (bounding boxÎ°ú cropÎêú Ïù¥ÎØ∏ÏßÄ Ïù¥Ïö©)

### üîÆObject keypoint estimation ÏïåÍ≥†Î¶¨Ï¶ò
: Í∞ùÏ≤¥ ÎÇ¥Î∂ÄÏùò Ï†êÏùÑ Ï∞æÎäî Í∏∞Ïà†
#### 1. top-down: bounding boxÎ•º Ï∞æÍ≥† box ÎÇ¥Î∂ÄÏùò keypoint ÏòàÏ∏°
#### 2. bottom-up: Ïù¥ÎØ∏ÏßÄ Ï†ÑÏ≤¥Ïùò keypointÎ•º Ï∞æÍ≥† point Í¥ÄÍ≥ÑÎ•º Ïù¥Ïö©Ìï¥ Íµ∞ÏßëÌôîÌï¥ÏÑú box ÏÉùÏÑ±
Ïö∞Î¶¨Îäî Ïù¥ Ï§ë top-down Î∞©ÏãùÏùÑ Ïù¥Ïö©ÌïúÎã§.

## üîÆDlib landmark localization
: ÏñºÍµ¥ÏóêÏÑú 68Í∞úÏùò Ïù¥Î™©Íµ¨ÎπÑ ÏúÑÏπò Ï∞æÍ∏∞

```python
# Ï†ÄÏû•Ìïú landmark Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
list_landmarks = []

# ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ÎßàÎã§ face landmarkÎ•º Ï∞æÏïÑÎÇ¥Í∏∞
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect) # (RGB, rectangle)
    # face landmark Ï¢åÌëú Ï†ÄÏû•ÌïòÍ∏∞
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

#### ÏÇ¨ÏßÑ ÌÅ¨Í∏∞ ÎïåÎ¨∏Ïóê Ï†êÏù¥ Ïûò Î≥¥Ïù¥ÏßÄ ÏïäÏïÑ ÌÅ¨Í∏∞Î•º 10ÏúºÎ°ú ÌÇ§ÏõåÏ£ºÏóàÎã§.

```python
# landmark Ï†ê Ï∞çÍ∏∞
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 10, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 4. Ïä§Ìã∞Ïª§ Ï†ÅÏö©ÌïòÍ∏∞
Ïö∞Î¶¨Îäî Í≥†ÏñëÏù¥ Ïä§Ìã∞Ïª§Î•º ÏΩîÏôÄ ÏûÖÏà† ÏÇ¨Ïù¥Ïóê Ï∞çÏñ¥Ï§Ñ Í≤ÉÏù¥Îã§!
#### ÏΩî ÎÅùÏùò indexÎäî 33Î≤à

```python
# Ï¢åÌëú ÌôïÏù∏ÌïòÍ∏∞

for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[33]) # ÏΩî ÎÅùÏùò index = 33
    x = landmark[33][0]
    y = landmark[33][1]
    w = h = dlib_rect.width()
    
    print('ÏΩîÏùò ÏúÑÏπò (x, y) : (%d, %d)' %(x, y))
    print('box ÌÅ¨Í∏∞ (w, h) : (%d, %d)' %(w, h))
```

##### Ïä§Ìã∞Ïª§Îäî png ÌååÏùºÎ°ú Î∂àÎü¨Ïò§Ïûê

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
img_sticker = cv2.resize(img_sticker, (w, h)) # Ïä§Ìã∞Ïª§ resize

print(img_sticker.shape)
plt.imshow(img_sticker)
```

### ‚úîÔ∏è Ïä§Ìã∞Ïª§Ïùò ÏãúÏûë Ï¢åÌëú Ïû°Í∏∞
#### Ïù¥ Ï¢åÌëúÍ≥ÑÎäî top-leftÏù¥Îã§. Îî∞ÎùºÏÑú ÏΩîÏùò ÏúÑÏπòÏóêÏÑú Î∞ïÏä§ Í∞ÄÎ°ú Í∏∏Ïù¥Ïùò Ï†àÎ∞òÎßåÌÅº ÏôºÏ™ΩÏúºÎ°ú Í∞ÄÏïºÌïòÍ≥†,
#### y Ï¢åÌëúÎäî (h//2Î°ú) Ï†ÅÎãπÌûà Ï°∞Ï†àÌïòÏòÄÎã§.

```python
# Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ ÏãúÏûë Ï¢åÌëú (top-left Ï¢åÌëú)

refined_x = x - w//2 # left
refined_y = y - h//2 # top
print('(x, y): (%d, %d)' %(refined_x, refined_y))
```

#### Ïä§Ìã∞Ïª§Í∞Ä ÏõêÎ≥∏ Ïù¥ÎØ∏Ïùò Î≤îÏúÑÎ•º Î≤óÏñ¥ÎÇòÎäî Í≤ΩÏö∞

```python
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print('(x, y): (%d, %d)' %(refined_x, refined_y))
```

```python
# img_show[from: to] ÌòïÏãù

# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Ïä§Ìã∞Ïª§ Ï†ÅÏö©Ìï† ÏúÑÏπò cropÌïú Ïù¥ÎØ∏ÏßÄ
sticker_area = img_show[refined_y: refined_y+img_sticker.shape[0],
                       refined_x: refined_x+img_sticker.shape[1]]
img_show[refined_y: refined_y+img_sticker.shape[0],
        refined_x: refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0, img_sticker, sticker_area).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

-> Ïä§Ìã∞Ïª§Í∞Ä Ïûò Î∂ÄÏ∞©ÎêòÏóàÎã§!

```python
# img_show ÎåÄÏã† img_rbg Ïù¥Ïö©Ìï¥Î≥¥Í∏∞

sticker_area = img_bgr[refined_y: refined_y+img_sticker.shape[0],
                       refined_x: refined_x+img_sticker.shape[1]]
img_bgr[refined_y: refined_y+img_sticker.shape[0],
        refined_x: refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0, img_sticker, sticker_area).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
