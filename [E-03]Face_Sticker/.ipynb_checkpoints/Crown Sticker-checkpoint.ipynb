# 3.3 ÏñºÍµ¥ Í≤ÄÏ∂ú Face Detection

```python
# detector ÏÑ†Ïñ∏
detector_hog = dlib.get_frontal_face_detector()
print("üå´üõ∏")
```

```python
# detector_hogÎ•º Ïù¥Ïö©Ìï¥ ÏñºÍµ¥Ïùò bounding box Ï∂îÏ∂ú

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1) # (image, # of image pyramid)
print("üå´üõ∏")
```

image pyramid
- Ïù¥ÎØ∏ÏßÄÎ•º upsampling Ìï¥ ÌÅ¨Í∏∞Î•º ÌÇ§Ïö∞Îäî Í≤É
- Ïù¥ÎØ∏ÏßÄ ÌîºÎùºÎØ∏ÎìúÏóêÏÑú ÏñºÍµ¥ÏùÑ Ïû¨Í≤ÄÏ∂úÌïòÎ©¥ ÏûëÍ≤å Ï¥¨ÏòÅÎêú ÏñºÍµ¥ÏùÑ ÌÅ¨Í≤å Î≥º Ïàò ÏûàÏùå

```python
# Ï∞æÏùÄ ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ Î¶¨Ïä§Ìä∏
# Ïó¨Îü¨ ÏñºÍµ¥Ïù¥ ÏûàÏùÑ Ïàò ÏûàÎã§.

print(dlib_rects)

for dlib_rect in dlib_rects:
    l = dlib_rect.left() # ÏôºÏ™Ω ÏïÑÎûò
    t = dlib_rect.top() # ÏôºÏ™Ω ÏúÑ
    r = dlib_rect.right() # Ïò§Î•∏Ï™Ω ÏúÑ
    b = dlib_rect.bottom() # Ïò§Î•∏Ï™Ω ÏïÑÎûò
    
    cv2.rectangle(img_show, (l, t), (r, b), (0, 255, 0), 2, lineType=cv2.LINE_AA)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 3.4 ÏñºÍµ¥ ÎûúÎìúÎßàÌÅ¨ Face Landmark

face landmark localization : Ïù¥Î™©Íµ¨ÎπÑÏùò ÏúÑÏπò Ï∂îÎ°†ÌïòÎäî Í∏∞Ïà†

## Dlib landmark localization
- ÏûòÎùºÏßÑ ÏñºÍµ¥ Ïù¥ÎØ∏ÏßÄÏóêÏÑú 68Í∞úÏùò Ïù¥Î™©Íµ¨ÎπÑ ÏúÑÏπò Ï∞æÍ∏∞
- dlibÏùÄ ÌïôÏäµÎêú pretrained modelÏùÑ Ï†úÍ≥µÌïúÎã§. Î®ºÏ†Ä Îã§Ïö¥Î∞õÏûê!

```python
# $ wget http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2
# $ mv shape_predictor_68_face_landmarks.dat.bz2 ~/aiffel/camera_sticker/models
# $ cd ~/aiffel/camera_sticker && bzip2 -d ./models/shape_predictor_68_face_landmarks.dat.bz2
```

```python
# Ï†ÄÏû•Ìïú landmark Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("üå´üõ∏")
```

landmark_predictorÏùÄ RGB Ïù¥ÎØ∏ÏßÄÏôÄ dlib.rectangleÏùÑ ÏûÖÎ†•Î∞õÏïÑ <br/>
dlib.full_object_detectionÏùÑ Î∞òÌôòÌï®

```python
list_landmarks = []

# ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ÎßàÎã§ face landmarkÎ•º Ï∞æÏïÑÎÇ¥Í∏∞
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect) # (RGB, rectangle)
    # face landmark Ï¢åÌëú Ï†ÄÏû•ÌïòÍ∏∞
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

```python
# landmark Ï†ê Ï∞çÍ∏∞
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 4, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

landmarkÍ∞Ä Ïù¥ÎØ∏ÏßÄÏóê Ï†ïÌôïÌûà Ï∂úÎ†•Îê® !!!

# 3-5. Ïä§Ìã∞Ïª§ Ï†ÅÏö©ÌïòÍ∏∞
- Ïö∞Î¶¨Îäî Î®∏Î¶¨Ïóê Ïä§Ìã∞Ïª§Î•º Î∂ôÏùº Í≤É
- ÏΩî Í∏∞Ï§Ä ÎÜíÏù¥ x Ïù¥ÏÉÅ / ÎààÏçπ ÏúÑ n ÌîΩÏÖÄ ÏúÑÏóê Íµ¨ÌòÑ Í∞ÄÎä•

```python
# Ï¢åÌëú ÌôïÏù∏ÌïòÍ∏∞

for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[30]) # ÏΩîÏùò index = 30
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.height()//2 # Ï¢åÏ∏° ÏÉÅÎã®ÏóêÏÑú ÏãúÏûëÌïòÎØÄÎ°ú ÎπºÍ∏∞
    w = h = dlib_rect.width()
    
    print('(x, y) : (%d, %d)' %(x, y))
    print('(w, h) : (%d, %d)' %(w, h))
```

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/king.png'
img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
img_sticker = cv2.resize(img_sticker, (w, h))
print(img_sticker.shape)
```

```python
# (ÏΩîÏùò) x, y Ï¢åÌëú Ï°∞Ï†ï (top-left Ï¢åÌëú)

refined_x = x - w//2
refined_y = y - h
print('(x, y): (%d, %d)' %(refined_x, refined_y))
```

- Ïä§Ìã∞Ïª§ ÏãúÏûë Ï†êÏù¥ ÏñºÍµ¥ ÏÇ¨ÏßÑ ÏòÅÏó≠ÏùÑ Î≤óÏñ¥ÎÇòÎ©¥ ÏùåÏàòÎ°ú ÌëúÌòÑÎê®
- ndarryÎäî ÏùåÏàò Ïù∏Îç±Ïä§Ïóê Ï†ëÍ∑ºÌï† Ïàò ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê ÏùåÏàòÏóê ÎåÄÌïú ÏòàÏô∏Ï≤òÎ¶¨Î•º Ìï¥Ï§òÏïº Ìï® -> Ï†úÍ±∞

```python
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print('(x, y): (%d, %d)' %(refined_x, refined_y))
```

```python
# img_show[from: to] ÌòïÏãù

# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Ïä§Ìã∞Ïª§ Ï†ÅÏö©Ìï† ÏúÑÏπò cropÌïú Ïù¥ÎØ∏ÏßÄ
sticker_area = img_show[refined_y: refined_y+img_sticker.shape[0],
                       refined_x: refined_x+img_sticker.shape[1]]
img_show[refined_y: refined_y+img_sticker.shape[0],
        refined_x: refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0, sticker_area, img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# img_show ÎåÄÏã† img_rbg Ïù¥Ïö©Ìï¥Î≥¥Í∏∞

sticker_area = img_bgr[refined_y: refined_y+img_sticker.shape[0],
                       refined_x: refined_x+img_sticker.shape[1]]
img_bgr[refined_y: refined_y+img_sticker.shape[0],
        refined_x: refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
